```python
import matplotlib.pyplot as plt 
import pandas as pd 
#Imports a CSV file to DataFrame format
data = pd.read_csv('popl_female.csv') 
  
df = pd.DataFrame(data) 
  
X = list(df.iloc[:, 0]) 
Y = list(df.iloc[:, 1]) 
  
# Plot the data using bar() method 
plt.bar(X, Y, color='g') 
plt.title("female Population(in billions)") 
plt.xlabel("Years") 
plt.ylabel("Population") 
  
# Show the plot 
plt.show() 
```
